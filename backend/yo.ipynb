```python
import requests, json
from collections import defaultdict

BASE = "https://gamma-api.polymarket.com/events"

liquidity_markets = defaultdict(int)
params = {
    "order": "volume",      # <-- use a real field
    "ascending": False,      # send a boolean, not a string
    "limit": 10,
    "closed": False, # only get open markets
    # "liquidity_num_min": 100000
}

r = requests.get(BASE, params=params, timeout=10)
r.raise_for_status()
events = r.json()
for e in events:
    liquidity_markets[e.get("question")] = e.get("liquidity_num")
    
print(events)
```
